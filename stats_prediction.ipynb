```python
def parse_ht(ht):
    # format: 7' 0.0"
    ht_ = ht.split("-")
    ft_ = float(ht_[0])
    in_ = float(ht_[1])
    return (12*ft_) + in_
    stats_df["height"] = stats_df["height"].apply(parse_ht(ht))

print (stats_df["height"].apply(lambda x:parse_ht(x)))
```

```python
stats_df["height"] = stats_df["height"].apply(lambda x:parse_ht(x))
stats_df.head()
```
