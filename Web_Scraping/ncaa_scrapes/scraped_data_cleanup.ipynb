```python
import pandas as pd
import string
```

```python
base_path = 'names_merge_stats/'
file_suffix = '_name_stats.csv'

alphabet = list(string.ascii_lowercase)
```

```python
# read in 'a' name players
df = pd.read_csv(base_path + 'a' + file_suffix)
df
```

```python
# iterate through all letters, adding new letter df to bottom of main df
for i in range(1,26):
    temp_df = pd.read_csv(base_path + alphabet[i] + file_suffix)
    df = pd.concat([df, temp_df], axis=0, ignore_index=True)
    
df
```

```python
# read in nba data
nba_df = pd.read_csv('namedLeaguePlayers_collegeURL.csv')

# pull ncaa urls into separate column
nba_college_urls = nba_df['college_url'].tolist()
nba_college_urls

# generate list of college urls short forms
short_urls_unfiltered = []
for x in range(0, len(nba_college_urls)):
    if pd.isna(nba_college_urls[x]):
        short_url = None
    else:
        short_url = nba_college_urls[x][32:]
    
    short_urls_unfiltered.append(short_url)

# filter None values from list
short_urls = list(filter(None, short_urls_unfiltered))
```

```python
# pull urls from ncaa player df
ncaa_url_list = df['url'].tolist()

# define list for whether ncaa player played professionally
is_pro_list = []

# iterate through all ncaa players checking if they made the pro list
for ncaa_player in ncaa_url_list:
    if ncaa_player in short_urls:
        is_pro_list.append(1)
    else:
        is_pro_list.append(0)
```

```python
# add is_pro column to df 
df['is_pro'] = is_pro_list

df
```

```python
# write combined df to a csv
df.to_csv('combined_ncaa_player_stats.csv', index=False)
```
